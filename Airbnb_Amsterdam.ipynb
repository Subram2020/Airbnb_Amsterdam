# Top 10 Values

```python
dfs = []

for col in df.columns:
    top_values = []
    if is_numeric_dtype(df[col]):
        top_values = df[col].nlargest(n=10)
        dfs.append(pd.DataFrame({col: top_values}).reset_index(drop=True))
pd.concat(dfs, axis=1)
```

# Top 10 Smallest Values

```python
dfs = []

for col in df.columns:
    smallest_values = []
    if is_numeric_dtype(df[col]):
         smallest_values = df[col].nsmallest(n=10)
         dfs.append(pd.DataFrame({col: smallest_values}).reset_index(drop=True))
pd.concat(dfs, axis=1)
```

# Converting date time into Year & Month 

```python
df['last_review']=pd.to_datetime(df['last_review'])
df['Mon_Year'] = df['last_review'].dt.strftime('%Y-%m')
```

```python
df.Mon_Year.head(10)
```

```python
last_rev = df.groupby(by = 'Mon_Year').agg({'Mon_Year':'count'}).sort_index().tail(36)
last_rev.tail(10)
```

```python
df['month'] = df['last_review'].dt.strftime('%b')
df.month.head(10)
```

```python
df['Year'] = df['last_review'].dt.strftime('%Y')
df.Year.head(10)
```

```python
months_in_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
review_count = df.groupby(by=["Year", "month"]).size().reset_index()
review_count.tail(10)
```

```python
review_count.columns
```

```python
review_count.columns = ['Year','month','reviews']
```

```python
reviews_pivoted = review_count[review_count['Year']>'2014'].pivot(index='month', columns='Year', values='reviews').reindex(months_in_order)
reviews_pivoted2 = reviews_pivoted.iloc[:,:-2]
reviews_pivoted2
```

# Reviews in 2015 & 2016

```python
reviews_pivoted2.plot(kind='bar', figsize=(15,8), color=['blue', 'green', 'orange', 'purple', 'black'])
plt.title("Count of reviews by month and year", y=1.013, fontsize=22)
plt.xlabel("Month", labelpad=16)
plt.ylabel("Count of reviews", labelpad=16);  
```
